```python
# Core set (your final choice)
core_features = [
    "chest_emg_median_freq",
    "chest_emg_mean_freq",
    "chest_emg_iemg",
    "chest_emg_rms",
    "chest_emg_burst_duration"
]

# Add nonlinear support
core_plus = core_features + ["chest_emg_total_power"]

# Add combined feature
final_set = core_plus + ["freq_rms_ratio"]

# Full (everything, for comparison)
all_features = [col for col in df.columns if "chest_emg" in col]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

model = RandomForestClassifier(n_estimators=100, random_state=42)

def evaluate(features):
    X = df[features]
    y = df["label"]
    scores = cross_val_score(model, X, y, cv=5)
    return scores.mean(), scores.std()
```

```python
print("Core:", evaluate(core_features))
print("Core + 1:", evaluate(core_plus))
print("Final:", evaluate(final_set))
print("All features:", evaluate(all_features))
```
